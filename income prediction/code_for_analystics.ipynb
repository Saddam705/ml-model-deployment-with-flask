```python
import numpy as np
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('adult.csv')
df.head()
```

```python
df.info()
```

```python
df = df.drop(['fnlwgt','educational-num'],axis=1)
df_columns = df.columns
```

```python
for c in df_columns:
    df = df.replace('?',np.NaN)
df = df.apply(lambda x: x.fillna(x.value_counts().index[0]))
```

```python
df.replace(['Divorced', 'Married-AF-spouse',
			'Married-civ-spouse', 'Married-spouse-absent',
			'Never-married', 'Separated', 'Widowed'],
		['divorced', 'married', 'married', 'married',
			'not married', 'not married', 'not married'], inplace=True)
```

```python
category_col = ['workclass', 'race', 'education', 'marital-status', 'occupation',
				'relationship', 'gender', 'native-country', 'income']
from sklearn.preprocessing  import LabelEncoder
from  sklearn.tree import DecisionTreeClassifier
```

```python
df.head()
```

```python
from sklearn.pipeline import  Pipeline,make_pipeline
from sklearn.compose import ColumnTransformer
prepro = Pipeline(steps=[('label_encoder',LabelEncoder())])
pipe = ColumnTransformer(transformers=[('column_transfer',prepro,['workclass', 'race', 'education', 'marital-status', 'occupation',
				'relationship', 'gender', 'native-country', 'income'])])
model = make_pipeline(pipe,DecisionTreeClassifier())
```

```python
model
```

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X = df.values[:,0:12]
Y = df.values[:,12]
```

```python
X_train, X_test, y_train,  y_test = train_test_split(X,Y,test_size=0.3,random_state=100)
model.fit(X_train,y_train)
# y_pred_gini = pipe.predict(X_test)
# print('Decision Tree using Gini index\nAccuracy is',accuracy_score(y_test,y_pred_gini)*100)
```
